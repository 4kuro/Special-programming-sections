<h3><b>Вступ.</b></h3><br>
<b>Anaconda</b> - це комплексне середовище для програмування, яке включає в себе інструменти для управління пакетами, віртуальних середовищ, розробки та виконання коду. Anaconda забезпечує зручний та уніфікований спосіб управління середовищами та ресурсами для розробки, дозволяючи розробникам ефективно працювати з великим обсягом коду та данних.<br>
<b>IPython</b> – середовище розробки яке розширює консольне представлення Python до інтерактивної та ітеративної розробки. IPython notebook використовується через веб-застосунок, в якому можна вести усі аспекти розробки та обчислень: програмування, ведення документації, виконання коду, збереження результатів виконання коду. <br>
<b>Jupyter</b> – середовище розробки, покращена версія IPython notebook, що підтримує десятки мов (наприклад Ruby, R, Haskell, і, звичайно, Python). Оскільки Jupyter notebook побудований на засадах Ipython, він має схожі, часто більш потужні, застосування. Він широко використовується в області машинного навчання, аналізу даних та інших галузях, де важливий взаємозв'язок між кодом, результатами та поясненнями.<br>
<b>Notebook (блокнот)</b> – файл IPython або Jupyter, що складається з клітинок (cells). Кожна клітинка може містити код, сирий текст, або ж текст з розміткою Markdown та LaTeX. Після кожної клітинки з кодом виводиться результат роботи клітинки: виводи команд print, графіки, зображення, т.і. <br>

Математичне сподівання та дисперсія σ даних

```python
titles = ['radius', 'texture', 'perimeter', 'area']

for i in titles:
    print(f'\nМатематичне сподівання {i} =', abs(round(df[f'{i}'].mean(), 3)))
    print(f'Дисперсія σ {i} =', round(df[f'{i}'].std(), 3))

print('\n')
```

Функція стандартизації обраних колонок у датафреймі:

```python
def standartize(df, columns):
    df_st = pd.DataFrame()
    for column in columns:
        mean = df[column].mean()
        std = df[column].std()
        df_st[f'{column}_normalized'] = (df[column] - mean) / std
    return df_st
```

Результати стандартизації:

```python
df_standartized = standartize(df, df.columns[1:])
#standartize(df, df.columns[1:])
```
